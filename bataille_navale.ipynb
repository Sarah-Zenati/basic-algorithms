## Bataille Navale 

```python
import random

# Créer la grille de jeu
def creer_grille():
    return [[' ']*10 for _ in range(10)]

# Afficher la grille
def afficher_grille(grille):
    print("  A B C D E F G H I J")
    print("---------------------")
    for i in range(10):
        print(str(i) + "|"+ "|".join(grille[i]) + "|")
        print("---------------------")

# Placer aléatoirement les bateaux
def placer_bateaux(grille):
    bateaux = {'Porte-avions': 5, 'Croiseur': 4, 'Destroyer': 3, 'Sous-marin': 3, 'Torpilleur': 2}
    for bateau, taille in bateaux.items():
        placer_bateau(grille, bateau, taille)

# Placer un bateau sur la grille
def placer_bateau(grille, bateau, taille):
    directions = [(1, 0), (0, 1)]
    while True:
        x = random.randint(0, 9)
        y = random.randint(0, 9)
        direction = random.choice(directions)
        dx, dy = direction
        if x + (taille - 1) * dx < 10 and y + (taille - 1) * dy < 10:
            if all(grille[x + i * dx][y + i * dy] == ' ' for i in range(taille)):
                for i in range(taille):
                    grille[x + i * dx][y + i * dy] = bateau[0]
                break

# Vérifier si une coordonnée est valide
def coordonnee_valide(x, y):
    return 0 <= x < 10 and 0 <= y < 10

# Demander au joueur des coordonnées
def demander_coordonnees():
    while True:
        try:
            coordonnees = input("Entrez les coordonnées (ex: A3): ").strip().upper()
            if len(coordonnees) != 2:
                raise ValueError
            colonne = ord(coordonnees[0]) - ord('A')
            ligne = int(coordonnees[1])
            if not coordonnee_valide(ligne, colonne):
                raise ValueError
            return ligne, colonne
        except ValueError:
            print("Coordonnées invalides. Assurez-vous d'entrer une lettre suivie d'un chiffre (ex: A3).")

# Jouer une partie de bataille navale
def jouer_bataille_navale():
    print("Bienvenue dans la Bataille Navale !")
    grille = creer_grille()
    placer_bateaux(grille)
    afficher_grille(grille)
    while True:
        ligne, colonne = demander_coordonnees()
        if grille[ligne][colonne] != ' ':
            print("Touché !")
            grille[ligne][colonne] = 'X'
            afficher_grille(grille)
        else:
            print("À l'eau !")
            grille[ligne][colonne] = '-'
            afficher_grille(grille)

# Lancer le jeu
if __name__ == "__main__":
    jouer_bataille_navale()
```
