```python
def sqr(x):
    y = x**2 # on élève x au carré
    return y # on renvoie y
```

We want to plot the square function defined above over the interval I = [-5,5].

First, we discretize I into a vector of N equidistributed elements, with the first element being -5 and the last one being 5.

```python
N =50
x = np.linspace(-5,  5, N)
print(x)
x.shape
```

Next, we evaluate the square function at every point in the vector x.


```python
y = sqr(x)
#print(y)
```

To plot the square function, we use the Pyplot library.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
plt.plot(x, y)

plt.figure(figsize = (5, 5))
plt.plot(x, y, linestyle = '--', color = 'red')
plt.xlim(-5,5)
plt.ylim(0,50)
plt.xlabel("La variable x")
plt.ylabel("La variable y")
plt.title("La fonction x -> x^2", color = 'blue', fontstyle = 'italic')
```
